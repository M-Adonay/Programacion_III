```python
import pandas as pd
import matplotlib.pyplot as Graficas

try:
  tarjetacredito = pd.read_csv("/content/trx_tarjetas_credito (3).csv",sep=",")
  print("Archivo Cargado con exito")
except:
  print('Por favor cargar el archivo:\n trx_tarjetas_credito.csv')
```

```python
creditoGender = pd.crosstab(index=tarjetacredito["Gender"],columns="Cuenta")
print(creditoGender)

pieGender = creditoGender.loc[creditoGender.index]["Cuenta"]
Graficas.pie(pieGender, labels=pieGender.index)
Graficas.title("Uso de la tarjeta por sexo")
Graficas.show()

creditoGender.to_excel("Tarjeta_por_sexo.xlsx")
```

```python
totalDlsCiudad = tarjetacredito.groupby(["City", "Card Type"]).sum()
print(totalDlsCiudad)

pieCity = totalDlsCiudad.loc[totalDlsCiudad[:10].index]
Graficas.pie(pieCity, labels=pieCity.index)
Graficas.title("Total de Dls y tarjeta, Primeras 10 ciudades")
Graficas.show()

totalDlsCiudad.to_excel("total_dls_ciudad_tipo_tarjeta.xlsx")
```

```python
totalDlsTipo = pd.crosstab(index=tarjetacredito["Exp Type"],columns="Gastos")
print(totalDlsTipo)

pieGastos = totalDlsTipo.loc[totalDlsTipo.index]["Gastos"]
Graficas.pie(pieGastos, labels=pieGastos.index)
Graficas.title("Total por tipos de gastos")
Graficas.show()

totalDlsTipo.to_excel("total_dls_tipo_gasto.xlsx")
```
